**Needed Data:**
To achieve this we need 
**1-** Metropolitan Transportation Authority (MTA) dataset [Link](http://web.mta.info/developers/turnstile.html) which provides the basic data about each turnstile (e.g. turnstile id, data observation date and time, etc) in addition to their quantitive entries and exits data. 
**2-** NYC Taxi datasert [Link](https://www1.nyc.gov/site/tlc/about/tlc-trip-record-data.pa) which provide additinal data about poeple transportation patterns. 
**3-** NYC Turnstiles Station List dataset [Link](https://qri.cloud/nyc-transit-data/turnstiles_station_list) which provides the needed spatial data to combine with the NYC taxi dataset.
